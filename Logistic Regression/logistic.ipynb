# Logistic Regression

```python
# Let's start applying the logistic regression

#X = data.drop(["diagnosis"],axis=1)
feature_columns = [col for col in data.columns if col != 'diagnosis']
X = data[feature_columns]
y = data["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=40)

# fit the model with data
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
# Problem Occurs because data is not normalized as some coloumn has values much greater and etc. 
# Sklearn provided the method to do the normalization
from sklearn.preprocessing import StandardScaler 
scaler  = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Now Apply Logistic Regression Again

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=40)

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=40)

# fit the model with data
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)  # Prediction upon the test data
```

# Evalution

```python
# we have functions from sklearn as well
from sklearn.metrics import accuracy_score
accracy = accuracy_score(y_test, y_pred)
print(f"Accuracy : {accracy: .2f}")
```

```python
# Another method
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
#Diagonal values represent accurate predictions, while non-diagonal elements are inaccurate predictions
cnf_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
